## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(columns=['EIN','NAME'])
application_df.head()
```

```python
# Determine the number of unique values in each column.
application_df.nunique()
```

```python
application_df.info()
```

```python
# Look at APPLICATION_TYPE value counts for binning
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
type_values = application_df['APPLICATION_TYPE'].value_counts()
application_types_to_replace = type_values[type_values <500].index.to_list()

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
print(application_df['CLASSIFICATION'].nunique())
application_df['CLASSIFICATION'].value_counts()
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
class_values = application_df['CLASSIFICATION'].value_counts()
classifications_to_replace = class_values[class_values > 1]
classifications_to_replace
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
class_values = application_df['CLASSIFICATION'].value_counts()
classifications_to_replace = class_values[class_values < 1800].index.to_list()
# type_values = application_df['APPLICATION_TYPE'].value_counts()
# application_types_to_replace = type_values[type_values <500]

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
application_df.info()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
columns_of_interest = ['APPLICATION_TYPE','AFFILIATION','CLASSIFICATION','USE_CASE','ORGANIZATION','INCOME_AMT','SPECIAL_CONSIDERATIONS']
encoded_items = pd.get_dummies(application_df[columns_of_interest])
encoded_df = pd.concat([application_df,encoded_items], axis=1)
encoded_df = encoded_df.drop(columns=columns_of_interest, axis=1)
encoded_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
y = encoded_df['IS_SUCCESSFUL']
X = encoded_df.drop(columns='IS_SUCCESSFUL')
# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the 1st model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
# Base model

nn1 = tf.keras.models.Sequential()
input_features = X_train.shape[1]

# First hidden layer
nn1.add(tf.keras.layers.Dense(units=80, activation='relu', input_dim=input_features))
# Second hidden layer
nn1.add(tf.keras.layers.Dense(units=30, activation='relu'))
# Output layer
nn1.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
# Check the structure of the model
nn1.summary()
```

```python
# Compile the model
nn1.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Train the model
fit_model1 = nn1.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn1.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Define the 2nd model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
# Changes made from 1st model: added another hidden layer with 20 units

nn2 = tf.keras.models.Sequential()
input_features = X_train.shape[1]

# First hidden layer
nn2.add(tf.keras.layers.Dense(units=80, activation='relu', input_dim=input_features))
# Second hidden layer
nn2.add(tf.keras.layers.Dense(units=50, activation='relu'))
# Third hidden layer
nn2.add(tf.keras.layers.Dense(units=30, activation='relu'))
# Output layer
nn2.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
# Check the structure of the model
nn2.summary()
```

```python
nn2.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
fit_model2 = nn2.fit(X_train_scaled, y_train, epochs=100)
```

```python
model_loss, model_accuracy = nn2.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Define the 3rd model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
# Changes made from second model: increased epochs by 50

nn3 = tf.keras.models.Sequential()
input_features = X_train.shape[1]

# First hidden layer
nn3.add(tf.keras.layers.Dense(units=80, activation='relu', input_dim=input_features))
# Second hidden layer
nn3.add(tf.keras.layers.Dense(units=50, activation='relu'))
# Third hidden layer
nn3.add(tf.keras.layers.Dense(units=30, activation='relu'))
# Output layer
nn3.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
# Check the structure of the model
nn3.summary()
```

```python
nn3.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
fit_model3 = nn3.fit(X_train_scaled, y_train, epochs=150)
```

```python
model_loss, model_accuracy = nn3.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
